# Limits of Real Functions

## 1. Definition of Limits

The formal definition of the limit of a function \( f(x) \) as \( x \to a \) is:

$$
\lim_{x \to a} f(x) = L
$$

This means that as \( x \) approaches \( a \), the values of \( f(x) \) approach \( L \).

## 2. Limit at Infinity

Limits at infinity describe the behavior of a function as \( x \to \infty \) or \( x \to -\infty \).

### Limit at Infinity:
$$
\lim_{x \to \infty} f(x) = L
$$

### Example:
$$
\lim_{x \to \infty} \frac{1}{x} = 0
$$

## 3. Types of Limits

### 3.1 Finite Limits
A finite limit occurs when \( f(x) \) approaches a finite value as \( x \to a \).

Example:
$$
\lim_{x \to 2} \frac{x^2 - 4}{x - 2} = 4
$$

### 3.2 Infinite Limits
Infinite limits describe when the function approaches infinity or negative infinity.

Example:
$$
\lim_{x \to 0} \frac{1}{x^2} = \infty
$$

### 3.3 Limits at Infinity (Function Behavior at Large \( x \))
These limits describe the behavior of \( f(x) \) as \( x \to \infty \).

Example:
$$
\lim_{x \to \infty} \frac{x}{x+1} = 1
$$

## 4. Squeeze Theorem

If \( g(x) \leq f(x) \leq h(x) \) and
$$
\lim_{x \to a} g(x) = \lim_{x \to a} h(x) = L
$$
then by the Squeeze Theorem:
$$
\lim_{x \to a} f(x) = L
$$

### Example:
$$
\lim_{x \to 0} \frac{\sin(x)}{x} = 1
$$

## 5. Common Limits

Here are a few standard limits to remember:

- $$ \lim_{x \to 0} \sin(x) = 0 $$
- $$ \lim_{x \to 0} \cos(x) = 1 $$
- $$ \lim_{x \to \infty} \frac{1}{x} = 0 $$
- $$ \lim_{x \to \infty} (1 + \frac{1}{x})^x = e $$

## 6. Techniques for Finding Limits

### 6.1 Direct Substitution
If the function is continuous at \( x = a \), simply substitute \( x = a \) into the function.

### 6.2 Factoring
Factor the expression and cancel out terms if possible.

Example:
$$
\lim_{x \to 2} \frac{x^2 - 4}{x - 2} = \lim_{x \to 2} \frac{(x - 2)(x + 2)}{x - 2} = 4
$$

### 6.3 Rationalizing
For expressions involving square roots, multiply by the conjugate to simplify.

Example:
$$
\lim_{x \to 0} \frac{\sqrt{x+1} - 1}{x} = \frac{1}{2}
$$

### 6.4 L'Hopital's Rule
For indeterminate forms like \( \frac{0}{0} \) or \( \frac{\infty}{\infty} \), differentiate the numerator and denominator:

$$
\lim_{x \to a} \frac{f(x)}{g(x)} = \lim_{x \to a} \frac{f'(x)}{g'(x)}
$$

### Example for L'Hopital's Rule:
$$
\lim_{x \to 0} \frac{\sin(x)}{x} = \lim_{x \to 0} \frac{\cos(x)}{1} = 1
$$

## 7. Indeterminate Forms

Some limits result in indeterminate forms like \( \frac{0}{0} \), \( \frac{\infty}{\infty} \), or \( 0 \times \infty \). These require further analysis, often using the techniques mentioned above.

### Common Indeterminate Forms:
- \( \frac{0}{0} \)
- \( \frac{\infty}{\infty} \)
- \( 0 \times \infty \)
- \( \infty - \infty \)
- \( 1^\infty \)
- \( 0^0 \)
- \( \infty^0 \)

## 8. Example Problems

### Example 1:
Find \( \lim_{x \to 3} \frac{x^2 - 9}{x - 3} \).

**Solution:**
Factor the numerator:
$$
\frac{x^2 - 9}{x - 3} = \frac{(x - 3)(x + 3)}{x - 3}
$$
Cancel out \( x - 3 \):
$$
\lim_{x \to 3} (x + 3) = 6
$$

### Example 2:
Find \( \lim_{x \to 0} \frac{\sin(x)}{x} \).

**Solution:**
Using the well-known limit:
$$
\lim_{x \to 0} \frac{\sin(x)}{x} = 1
$$

```python
import sympy as sp

# Define the variable
x = sp.symbols('x')

# Define the function
f = (x**3 + 2*x**2) / (x**4 - 3*x**3)

# Compute the limit as x approaches infinity
limit_value = sp.limit(f, x, sp.oo)

# Print the result
print(f"The limit as x approaches infinity is: {limit_value}")

```

```python
import sympy as sp
import matplotlib.pyplot as plt
import numpy as np

# Define the variables
x = sp.symbols('x')

# Define the functions
f = (x**2 - 1) / (x**2 + 1)
g = sp.sin(x) / (x**2 + 1)

# Compute limits at infinity for horizontal asymptotes
lim_f_pos_inf = sp.limit(f, x, sp.oo)
lim_f_neg_inf = sp.limit(f, x, -sp.oo)

lim_g_pos_inf = sp.limit(g, x, sp.oo)
lim_g_neg_inf = sp.limit(g, x, -sp.oo)

# Print limits (horizontal asymptotes)
print(f"f(x) horizontal asymptote: {lim_f_pos_inf} (as x -> ±∞)")
print(f"g(x) horizontal asymptote: {lim_g_pos_inf} (as x -> ±∞)")

# Plotting functions
x_vals = np.linspace(-10, 10, 400)
f_vals = (x_vals**2 - 1) / (x_vals**2 + 1)
g_vals = np.sin(x_vals) / (x_vals**2 + 1)

plt.plot(x_vals, f_vals, label=r"$f(x) = \frac{x^2 - 1}{x^2 + 1}$")
plt.axhline(y=1, color='r', linestyle='--', label='y=1 (horizontal asymptote)')
plt.axhline(y=0, color='g', linestyle='--', label='y=0 (horizontal asymptote)')

plt.plot(x_vals, g_vals, label=r"$g(x) = \frac{\sin(x)}{x^2 + 1}$")
plt.title("Asymptotes of f(x) and g(x)")
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```
