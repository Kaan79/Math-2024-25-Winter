# General Equations

## Ellipsoid
The general equation of an ellipsoid is:

$$
\frac{x^2}{a^2} + \frac{y^2}{b^2} + \frac{z^2}{c^2} = 1
$$

Where:
- $a$, $b$, and $c$ are the semi-axes of the ellipsoid along the $x$, $y$, and $z$ axes.

## Sphere
The general equation of a sphere is:

$$
x^2 + y^2 + z^2 = r^2
$$

Where:
- $r$ is the radius of the sphere.

## Hyperboloid
The general equation of a hyperboloid is:

$$
\frac{x^2}{a^2} - \frac{y^2}{b^2} - \frac{z^2}{c^2} = 1
$$

Where:
- $a$, $b$, and $c$ define the scale along the respective axes.

# Rotational Surfaces

A surface of revolution is formed by rotating a curve around an axis. For example, rotating a function $f(x)$ around the $x$-axis forms a rotational surface. The equation of the surface can be expressed in parametric form as:

$$
x = u
$$

$$
y = f(u) \cdot \cos(v)
$$

$$
z = f(u) \cdot \sin(v)
$$

Where:
- $u$ is the parameter along the axis of rotation (e.g., the $x$-axis).
- $v$ is the parameter for the angle of rotation (from $0$ to $2\pi$).

```python
import sympy as sp

# Define the symbols
x, y, z = sp.symbols('x y z')

# Center of the sphere and radius
center = (1, 2, 3)
radius = 3

# Equation of the sphere
sphere_eq = sp.Eq((x - center[0])**2 + (y - center[1])**2 + (z - center[2])**2, radius**2)

# Display the equation of the sphere
sphere_eq
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the sphere equation
def sphere(radius):
    u = np.linspace(0, 2 * np.pi, 100)
    v = np.linspace(0, np.pi, 100)
    x = radius * np.outer(np.cos(u), np.sin(v))
    y = radius * np.outer(np.sin(u), np.sin(v))
    z = radius * np.outer(np.ones(np.size(u)), np.cos(v))
    return x, y, z

# Create figure and 3D axis
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot both spheres
x1, y1, z1 = sphere(1)  # First sphere with radius 1
x2, y2, z2 = sphere(np.sqrt(2))  # Second sphere with radius sqrt(2)

# Plotting the spheres
ax.plot_surface(x1, y1, z1, color='blue', alpha=0.5)
ax.plot_surface(x2, y2, z2, color='red', alpha=0.5)

# Labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Two Spheres: x^2 + y^2 + z^2 = 1 and x^2 + y^2 + z^2 = 2')

# Show plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the circle in the yz-plane
theta = np.linspace(0, 2 * np.pi, 100)
y = (np.sqrt(3)/2) * np.cos(theta)
z = (np.sqrt(3)/2) * np.sin(theta)

# Define x as a constant
x = np.full_like(y, 1/2)

# Create figure and 3D axis
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the curve of intersection
ax.plot(x, y, z, color='blue')

# Labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Curve of Intersection of Two Spheres')

# Show plot
plt.show()
```

```python
import sympy as sp

# Define symbols
x, y = sp.symbols('x y')

# Define the equation of the paraboloid
z = (x - 1)**2 + y**2 + 1

# Compute the partial derivatives
z_x = sp.diff(z, x)  # Partial derivative with respect to x
z_y = sp.diff(z, y)  # Partial derivative with respect to y

# Point P(1, 0, 1)
x0, y0, z0 = 1, 0, 1

# Evaluate the partial derivatives at point P
z_x_at_P = z_x.subs({x: x0, y: y0})
z_y_at_P = z_y.subs({x: x0, y: y0})

# Equation of the tangent plane: z - z0 = f_x(x0, y0) * (x - x0) + f_y(x0, y0) * (y - y0)
tangent_plane = z0 + z_x_at_P * (x - x0) + z_y_at_P * (y - y0)

# Display the result
tangent_plane.simplify()
```
